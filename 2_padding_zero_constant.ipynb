# **Padding**

Padding is the process of adding extra pixels to an image or array at the boundaries in order to apply a filter or operation on the edge pixels. Padding is necessary to avoid losing information at the edges of the image or array during the filtering process.

When an image or array is filtered, the filter kernel moves across the image or array, and the output value at each pixel is calculated based on the values of the neighboring pixels within the kernel window. However, for pixels near the edge of the image or array, some of the neighboring pixels may not be within the image or array boundaries. Padding can be used to extend the image or array boundaries so that all the pixels have a complete set of neighboring pixels within the kernel window.

Common padding techniques include:

**Zero padding:** extra pixels are added with a value of zero.

**Reflection padding:** extra pixels are added by reflecting the image or array at the edges.

**Replication padding:** extra pixels are added by replicating the value of the nearest pixel at the edges.

**Wrap padding:** extra pixels are added by wrapping the image or array around to the opposite edge.

# **Zero Padding**

image.png

Zero padding is a technique used in image processing to add extra pixels of value zero around the edges of an image. This is often done before applying certain types of image processing filters, such as convolutional filters, to prevent the filter from shifting the image and to avoid edge artifacts.

Here's an example of how to apply zero padding to an image in Python:

```python
import numpy as np
from PIL import Image

# Open an image
image = Image.open("arif1.jpg")

# Convert the image to a NumPy array
image_array = np.array(image)

# Define the amount of padding to add
pad_amount = 10

# Pad the image with zeros
padded_array = np.pad(image_array, ((pad_amount, pad_amount), (pad_amount, pad_amount), (0, 0)), mode="constant")

# Convert the padded array back to an image
padded_image = Image.fromarray(padded_array)

# Display the padded image
padded_image.show()
```

In this example, we open an image using the **Image class** from the PIL library, and then convert it to a NumPy array using the **np.array()** function. 

We then define the amount of padding to add around the edges of the image, in this case **10 pixels**.

We use the **np.pad()** function to add the padding, passing in the input array, the amount of padding to add on each side, and the mode of the padding (in this case, "constant" padding with zeros). 

We then convert the padded array back to an image using the **Image.fromarray()** method, and display the padded image using the **show()** method.

Note that in the **np.pad()** function call, we pass a tuple containing three tuples of padding amounts, one for each dimension of the array. 

The first two tuples represent the padding to add to the height and width dimensions of the image, respectively. 

The third tuple represents the padding to add to the color channel dimension of the image, which is zero in the case of a grayscale image and 3 in the case of an RGB image. 

In this example, we assume that the input image is an RGB image, so we pad with a tuple of (0, 0) for the color channel dimension.

**Pad the image with zeros**

`padded_array = np.pad(image_array, ((pad_amount, pad_amount), (pad_amount, pad_amount), (0, 0)), mode="constant")`

This code snippet uses NumPy's pad function to add padding of zeros to a 3-dimensional image_array. The padding is added to both sides of the 1st and 2nd dimensions (rows and columns) and no padding is added to the 3rd dimension (channels).

The `((pad_amount, pad_amount), (pad_amount, pad_amount), (0, 0))` argument specifies the amount of padding to add to each dimension. The first tuple specifies the padding amount for the 1st and 2nd dimensions, and the second tuple specifies the padding amount for the 3rd dimension (which is 0, indicating no padding).

The `mode="constant"` argument specifies that the padding should be filled with a constant value, which is 0 in this case.

Overall, this code is useful when dealing with image processing tasks such as convolutional neural networks, where adding padding to the edges of the image can help preserve the size of the output after convolutions.

**Convert the padded array back to an image**

`padded_image = Image.fromarray(padded_array)`

This code snippet converts the NumPy array padded_array back to an image using the **Image.fromarray()** function from the Python Imaging Library (PIL).

The padded_array is a NumPy array that was padded with zeros in the previous code snippet. It could have been padded for a variety of reasons, such as to increase the size of the image or to make it compatible with a specific image processing algorithm.

The **Image.fromarray()** function takes a NumPy array as input and returns a PIL image. The resulting padded_image can be used for further image processing or visualized using a Python image display library such as Matplotlib or OpenCV.

It's important to note that the NumPy array needs to be in a specific format for this function to work correctly. Specifically, it needs to have the shape (height, width, channels) and the data type needs to be uint8. If the array is in a different format, it may need to be converted before using this function.

# **Constant Padding**

image.png

To pad an image with a constant value using Python, you can use the **numpy.pad()** function from the NumPy library.

Here's an example code snippet:

```python
import numpy as np
from PIL import Image

# load the image
img = Image.open('arif2.jpg')

# convert to numpy array
img_array = np.array(img)

# specify the amount of padding on each side
pad_width = ((10, 10), (10, 10), (0, 0))  # (top, bottom), (left, right), (channel)

# pad the image with a constant value of 155
padded_img_array = np.pad(img_array, pad_width, mode='constant', constant_values=155)

# convert back to image format
padded_img = Image.fromarray(padded_img_array)

# save the padded image
padded_img.save('padded_image.jpg')
padded_img
```

In this example, **pad_width** specifies the amount of padding to add to the top, bottom, left, and right sides of the image. The (0, 0) value for the channel axis means that we are not padding that axis.

The mode argument specifies how to handle the padding. In this case, we are using the '**constant**' mode, which means that we will pad the image with a constant value.

The constant_values argument specifies the constant value to use for padding. In this case, we are using a value of 155.
