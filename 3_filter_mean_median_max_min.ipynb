# **Mean Filter**

A mean filter, also known as a box filter, is a linear filter that replaces each pixel in an image with the average of its surrounding pixels within a given neighborhood. This operation is often used to smooth an image and reduce noise.

**Read the image**

`img = imageio.imread('arif_noise.png')`

This code reads an image file named 'arif_noise.png' using the imread function from the imageio library. 

The image file is presumably located in the same directory as the Python script or Jupyter notebook where this code is executed.

**Apply the mean filter with a 3x3 kernel**

`filtered_img = filters.uniform_filter(img, size=3)`

This code applies a mean filter to the input image img using the **uniform_filter** function from the `scipy.ndimage.filters` module. The size parameter specifies the size of the kernel used for the mean filtering, which in this case is 3x3.

**Show the filtered image**

`plt.imshow(filtered_img, cmap='gray')`

Note that the **imshow** function from Matplotlib automatically scales the pixel values so that they are between 0 and 1, and maps them to colors using a colormap. If you want to display the image without any scaling or color mapping, you can use the '**gray**' colormap.

```python
import numpy as np
import scipy.ndimage.filters as filters
import imageio
import matplotlib.pyplot as plt

# Read the image
img = imageio.imread('arif_noise.png')

# Apply the mean filter with a 3x3 kernel
filtered_img = filters.uniform_filter(img, size=3)

# Show the filtered image
plt.imshow(filtered_img, cmap='gray')

# Display original and filtered images side by side
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(filtered_img),plt.title('Filtered')
plt.xticks([]), plt.yticks([])
plt.show()

# Save the filtered image
imageio.imwrite('filtered_image.png', filtered_img)
```

**Removing tick labels**

`plt.xticks([]), plt.yticks([])`

**plt.xticks([]) and plt.yticks([])** are Matplotlib functions that remove the ticks (i.e., the numbers or marks) from the x and y axes of a plot. By calling these functions with an empty list as argument, we effectively remove the tick labels and tick marks from the plot.

# **Median Filter**

A median filter is a nonlinear filter that replaces each pixel in an image with the median value of its surrounding pixels within a given neighborhood. This operation is often used to reduce noise in an image while preserving edges and fine details.

**Apply median filter with kernel size of 5x5**

`filtered_img = cv2.medianBlur(img, 5)`

This line applies a median filter to the input image using OpenCV's **cv2**.medianBlur function, with a kernel size of 5x5.

**cv2.medianBlur** is an OpenCV function that applies a median filter to an image.

**img** is the input image to be filtered.

5 is the size of the kernel used for the median filter. A kernel of size 5x5 means that the median value for each pixel is calculated based on the 5x5 neighborhood around that pixel.

The output of the **cv2.medianBlur** function is a filtered image with reduced noise. The filtered image is assigned to the variable **filtered_img**.

Note that the median filter is a nonlinear filter that replaces each pixel value with the median value of its surrounding pixels. 

This makes it effective at removing **salt-and-pepper noise** (i.e., randomly scattered bright or dark pixels) from images, while preserving the edges and details of the image. 

The **kernel size** determines the amount of smoothing applied to the image, with larger kernel sizes resulting in more smoothing.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
img = cv2.imread('arif_noise.png')

# Apply median filter with kernel size of 5x5
filtered_img = cv2.medianBlur(img, 5)

# Display original and filtered images side by side
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(filtered_img),plt.title('Filtered')
plt.xticks([]), plt.yticks([])
plt.show()
```

# **Max Filter**

A max filter is a nonlinear filter that replaces each pixel in an image with the maximum value of its surrounding pixels within a given neighborhood. This operation is often used to perform operations such as edge detection and morphological transformations, such as dilation.

In image processing, a max filter is a dilation operation where the value of each pixel in the output image is the maximum value of the corresponding pixels in the input image that lie within a structuring element or kernel.

**Apply a maximum filter with a 3x3 kernel**

`kernel_size = 3`
`filtered_img = cv2.dilate(img, np.ones((kernel_size,kernel_size), np.uint8))`

This code applies a maximum filter to the input image using OpenCV's **cv2.dilate** function, with a kernel size of 3x3.

**kernel_size** is a variable that specifies the size of the kernel used for the maximum filter. In this case, the kernel size is 3x3, meaning that the maximum value for each pixel is calculated based on the 3x3 neighborhood around that pixel.

`np.ones((kernel_size,kernel_size), np.uint8)` creates a 3x3 kernel of ones with data type np.uint8. This kernel is used to perform the maximum filter.

**cv2.dilate** is an OpenCV function that performs dilation, which is a mathematical operation that expands the bright regions of an image. 

When used with a kernel of ones, cv2.dilate effectively performs a maximum filter, replacing each pixel value with the maximum value of its surrounding pixels within the kernel neighborhood.

The output of the **cv2.dilate** function is a filtered image with increased brightness in the brighter regions of the image. The filtered image is assigned to the variable **filtered_img**.

Note that the maximum filter is a nonlinear filter that replaces each pixel value with the maximum value of its surrounding pixels within a given neighborhood. 

This makes it effective at enhancing bright regions of an image, while preserving the edges and details of the image. 

The kernel size determines the amount of smoothing applied to the image, with larger kernel sizes resulting in more smoothing.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('arif_noise.png')

# Apply a maximum filter with a 3x3 kernel
kernel_size = 3
filtered_img = cv2.dilate(img, np.ones((kernel_size,kernel_size), np.uint8))

# Display the original and filtered images side by side
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original')

plt.subplot(122)
plt.imshow(cv2.cvtColor(filtered_img, cv2.COLOR_BGR2RGB))
plt.title('Filtered')

plt.show()
```

`plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))`

This code displays an image using Matplotlib's **imshow** function, after converting the image from BGR to RGB color format using OpenCV's cv2.cvtColor function.

**cv2.cvtColor** is an OpenCV function that converts an image from one color space to another. In this case, **cv2.COLOR_BGR2RGB** is used to convert the input image from BGR format (the default format used by OpenCV) to RGB format (the format used by Matplotlib).

**plt.imshow** is a Matplotlib function that displays an image. The input to imshow is the converted image.

By default, imshow displays the image with the origin (0,0) at the top left corner of the plot. To change this behavior, you can use `plt.gca().invert_yaxis()` to invert the y-axis of the plot, so that the origin is at the bottom left corner of the plot. You can also use `plt.xticks([])` and `plt.yticks([])` to remove the tick labels from the plot.

**Here's what each part of the code does:**

`cv2.imread` loads an image from file in BGR format.

`cv2.dilate` applies a maximum filter to the image using a kernel_size by kernel_size square kernel of ones. 

The `np.ones` function creates a kernel of ones with the specified size, and the np.uint8 data type specifies that the kernel elements are 8-bit unsigned integers.

The `plt.subplot` and `plt.imshow` functions are used to display the original and filtered images side by side. 

The `cv2.cvtColor` function is used to convert the BGR format to RGB format that Matplotlib expects.

The `plt.title` function adds titles to the subplots.

You can adjust the **kernel_size** variable to change the size of the maximum filter kernel. Larger kernels will produce more smoothing and larger regions of uniform brightness.

# **Min Filter**

A min filter is a nonlinear filter that replaces each pixel in an image with the minimum value of its surrounding pixels within a given neighborhood. This operation is often used to perform operations such as edge detection and morphological transformations, such as erosion.

In image processing, a min filter is an erosion operation where the value of each pixel in the output image is the minimum value of the corresponding pixels in the input image that lie within a structuring element or kernel.

There are different ways to implement a min filter in Python, but one common method is to use OpenCV's cv2.erode function with a kernel of ones. Here's an example code that applies a min filter to an image using this method:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('arif_noise.png')

# Apply a min filter with a 3x3 kernel
kernel_size = 3
kernel = np.ones((kernel_size, kernel_size), np.uint8)
filtered_img = cv2.erode(img, kernel)

# Display the original and filtered images
plt.subplot(121), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(filtered_img, cv2.COLOR_BGR2RGB)), plt.title('Min Filtered')
plt.xticks([]), plt.yticks([])
plt.show()
```
